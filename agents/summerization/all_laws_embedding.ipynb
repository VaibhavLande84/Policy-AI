```python
from langchain_groq import ChatGroq

def extract_and_summarize(path):
    # 1. Extract text using OCR.extract_text
    print(f"Extracting text from {path}...")
    text_string = OCR.extract_text(path)
    
    # 2. Summarize the text using Groq
    print("Summarizing text using Groq...")
    model = ChatGroq(model="llama-3.3-70b-versatile")
    prompt = f"Please provide a comprehensive summary of the following document text:\n\n{text_string}"
    response = model.invoke(prompt)
    summary = response.content
    
    # 3. Save to MongoDB
    print("Saving summary to MongoDB...")
    db = client.local  # Using the 'local' database from earlier cells
    collection = db.pdf_summaries
    
    document = {
        "path": path,
        "summary": summary
    }
    
    result = collection.insert_one(document)
    print(f"Successfully saved summary with id {result.inserted_id}")
    
    return document
```
